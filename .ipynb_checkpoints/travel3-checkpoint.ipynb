```python
##lets fo featrure engineering
X=df.drop(columns='ProdTaken',axis=1)
y=df['ProdTaken']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
X.TypeofContact.unique()
X.Occupation.unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
```

```python
ohe=OneHotEncoder(drop='first')
ss=StandardScaler()
cat_features=X.select_dtypes(include='object').columns
num_features=X.select_dtypes(exclude='object').columns
preprocessor=ColumnTransformer([
    ('OneHotEncoder',ohe,cat_features),
    ('StandardScaler',ss,num_features)
])
```

```python
X_train=preprocessor.fit_transform(X_train)
X_test=preprocessor.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,precision_score,recall_score,f1_score,roc_auc_score,roc_curve
```

```python
models={
    'Random':RandomForestClassifier(),
   "DT":DecisionTreeClassifier(),
    "LG":LogisticRegression(),
    "ADA":AdaBoostClassifier(),
"GBC":GradientBoostingClassifier(),
    "XGBC":XGBClassifier()    
}
for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train,y_train)
    y_pred_test=model.predict(X_test)
    y_pred_train=model.predict(X_train)
    print("model performance for train data set")
    acc_score_train=accuracy_score(y_train,y_pred_train)
    confusion_score_train=confusion_matrix(y_train,y_pred_train)
    classification_score_train=classification_report(y_train,y_pred_train)
    precision_score_train=precision_score(y_train,y_pred_train)
    recall_score_train=recall_score(y_train,y_pred_train)
    f1_score_score_train=recall_score(y_train,y_pred_train)
    roc_train=roc_auc_score(y_train,y_pred_train)
    print("model performance for test ")
    acc_score_test=accuracy_score(y_test,y_pred_test)
    pre_score_test=precision_score(y_test,y_pred_test)
    cla_score_test=accuracy_score(y_test,y_pred_test)
    classification_score_test=classification_report(y_test,y_pred_test)
    f1_score_test=f1_score(y_test,y_pred_test)
    confusion_matrix_test=confusion_matrix(y_test,y_pred_test)
    recall_score_test=recall_score(y_test,y_pred_test)
    roc_test=roc_auc_score(y_test,y_pred_test)
    print(list(models.keys())[i])
    print('Model performance for Training set')
    print("- Accuracy: {:.4f}".format(acc_score_train))
    print((confusion_score_train))
    
    print((classification_score_train))
    print('- precision: {:.4f}'.format(precision_score_train))
    print('- recall: {:.4f}'.format(recall_score_train))
    print('- Precision: {:.4f}'.format(f1_score_score_train))
    print('- precision: {:.4f}'.format(roc_train))
    print('----------------------------------')
    print('Model performance for Test set')
    print((acc_score_test))
    print('- confusion_matrix: {:.4f}',(confusion_matrix_test))
    print('- classification: {:.4f}',(classification_score_test))
    print('- precision: {:.4f}'.format(pre_score_test))
    print('- recall: {:.4f}'.format(recall_score_test))
    print('- f1: {:.4f}'.format(f1_score_test))
    print('- Roc Auc Score: {:.4f}'.format(roc_test))

    
    print('='*35)
    print('\n')
    

    
    
```

```python
model_params={}
for name, model,params in randomsearchcv_model:
    random=RandomizedSearchCV(estimator=model,
                             param_distributions=params, n_iter=100, cv=3, verbose=2, n_jobs=-1)
    random.fit(X_train,y_train)
    model_params[name]=random.best_params_
    for model_name in model_params:
        print(model_params[model_name])
```

```python
## Plot ROC AUC Curve
from sklearn.metrics import roc_auc_score,roc_curve
plt.figure()

# Add the models to the list that you want to view on the ROC plot
auc_models = [
{
    'label': 'XGBClassifier',
    'model':XGBClassifier(n_estimators=200,max_depth=12,learning_rate=0.1,colsample_bytree=1 ),
    'auc':  0.8882
},
]
for algo in auc_models:
    model = algo['model'] 
    model.fit(X_train, y_train) 
    fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(X_test)[:,1])
# Calculate Area under the curve to display on the plot
    plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % (algo['label'], algo['auc']))
# Custom settings for the plot 
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('1-Specificity(False Positive Rate)')
plt.ylabel('Sensitivity(True Positive Rate)')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.savefig("auc.png")
plt.show() 
```
