```python
import pickle
with open('random_forest_model.pkl', 'rb') as file:
    Random_Forest=pickle.load(file)
with open('xgb_model.pkl', 'rb') as file:
    XGBClassifier=pickle.load(file)
with open('preprocessor.pkl', 'rb') as file:
    preprocessor=pickle.load(file)
```

```python

input_dict = {
    'Age': [38],
    'TypeofContact': ['Self Enquiry'],
    'CityTier': [1],
    'DurationOfPitch': [9],
    'Occupation': ['Large Business'],
    'Gender': ['Female'],
    'NumberOfFollowups': [3],
    'ProductPitched': ['Deluxe'],
    'PreferredPropertyStar': [3],
    'MaritalStatus': ['Unmarried'],
    'NumberOfTrips': [4],
    'Passport': [1],
    'PitchSatisfactionScore': [5],
    'OwnCar': [0],
    'Designation': ['Manager'],
    'MonthlyIncome': [1000],
    'TotalVisiting': [2]
}

input_df = pd.DataFrame(input_dict)
```

```python
X_input_transformed = preprocessor.transform(input_df)
```

```python
rf_pred =Random_Forest.predict(X_input_transformed)
rf_pred_proba =Random_Forest.predict_proba(X_input_transformed)[:, 1]
print("Random Forest Prediction:", rf_pred[0])
print("Random Forest Probability: chance that the customer will take the product", rf_pred_proba[0])
```
